```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Grau

```python
# Lê os dois arquivos
df_tsg_degree = pd.read_csv("/content/TSG_degree.csv")
df_onco_degree = pd.read_csv("/content/onc_degree.csv")
```

```python


# Adiciona uma coluna indicando o tipo
df_tsg_degree["Tipo"] = "TSG"
df_onco_degree["Tipo"] = "Oncogene"

# Junta os dois DataFrames
df_todos = pd.concat([df_tsg_degree, df_onco_degree])

sns.stripplot(data=df_todos, x="Tipo", y="Grau", jitter=True, size = 4)

plt.ylim(0,500)
plt.yticks(range(0,500,50))
plt.title("Distribuição do Grau: TSGs vs Oncogenes")
plt.ylabel("Grau")
plt.xlabel("")
plt.tight_layout()
plt.show()
```

```python

# Adiciona uma coluna indicando o tipo
df_tsg_degree["Tipo"] = "TSG"
df_onco_degree["Tipo"] = "Oncogene"

# Junta os dois DataFrames
df_todos = pd.concat([df_tsg_degree, df_onco_degree])

sns.boxplot(data=df_todos, x="Tipo", y="Grau")

plt.ylim(0,500)
plt.yticks(range(0,500,50))
plt.title("Distribuição do Grau: TSGs vs Oncogenes")
plt.ylabel("Grau")
plt.xlabel("")
plt.tight_layout()
plt.show()
```

```python

# Adiciona uma coluna indicando o tipo
df_tsg_degree["Tipo"] = "TSG"
df_onco_degree["Tipo"] = "Oncogene"

# Junta os dois DataFrames
df_degree = pd.concat([df_tsg_degree, df_onco_degree])

sns.violinplot(data=df_degree, x="Tipo", y="Grau")

plt.ylim(0,500)
plt.yticks(range(0,500,50))
plt.title("Distribuição do Grau: TSGs vs Oncogenes")
plt.ylabel("Grau")
plt.xlabel("")
plt.tight_layout()
plt.show()
```

# Tabela de Medidas de Centralidade

```python
# Lê os arquivos (sem cabeçalho)
# Calcula estatísticas
tabela = pd.DataFrame({
    "Grau TSG": [
        df_tsg_degree["Grau"].mean(),
        df_tsg_degree["Grau"].median(),
        df_tsg_degree["Grau"].std()
    ],
    "Grau Oncogene": [
        df_onco_degree["Grau"].mean(),
        df_onco_degree["Grau"].median(),
        df_onco_degree["Grau"].std()
    ],
    "Closeness TSG": [
        df_tsg_clos["Closeness"].mean(),
        df_tsg_clos["Closeness"].median(),
        df_tsg_clos["Closeness"].std()
    ],
    "Closeness Oncogene": [
        df_onco_clos["Closeness"].mean(),
        df_onco_clos["Closeness"].median(),
        df_onco_clos["Closeness"].std()
    ],
    "Eigenvector TSG": [
        df_tsg_eigen["Eigenvector"].mean(),
        df_tsg_eigen["Eigenvector"].median(),
        df_tsg_eigen["Eigenvector"].std()
    ],
    "Eigenvector Oncogene": [
        df_onco_eigen["Eigenvector"].mean(),
        df_onco_eigen["Eigenvector"].median(),
        df_onco_eigen["Eigenvector"].std()
    ],
    "Betweenness TSG": [
        df_tsg_betw["Betweenness"].mean(),
        df_tsg_betw["Betweenness"].median(),
        df_tsg_betw["Betweenness"].std()
    ],
    "Betweenness Oncogene": [
        df_onco_betw["Betweenness"].mean(),
        df_onco_betw["Betweenness"].median(),
        df_onco_betw["Betweenness"].std()
    ]
}, index=["Média", "Mediana", "Desvio Padrão"])

tabela.to_csv("estatisticas.csv")

# Mostra a tabela
display(tabela)
```
